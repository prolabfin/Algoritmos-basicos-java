```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")
x = penguins["flipper_length_mm"]
y =  penguins["body_mass_g"]

fig, ax = plt.subplots()

# Graficar los datos
ax.plot(x, y, color='tab:red', label='Peso corporal')
ax.plot(x, x**2, color='tab:blue', label='Cuadrado de la longitud de la aleta')

# Etiquetas de los ejes y título del gráfico
ax.set_xlabel('Longitud de la aleta (mm)')
ax.set_ylabel('Peso corporal (g)')
ax.set_title('Relación entre la longitud de la aleta y el peso corporal de los pingüinos')

# Leyenda
ax.legend()

plt.show()
```
